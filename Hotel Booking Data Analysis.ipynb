# Exploratory Data Analysis

```python
#first we'll see the count of reservations that are cancelled and that are not.
cancelled_percent=df['is_canceled'].value_counts(normalize=True) *100
print(cancelled_percent)
#value_counts() returns the category name and its respective count.
#the normalize=True parameter returns the percentage
```

```python
#from the above output we can conclude that 62% reservations are not cancelled 
#and 37% reservations are cancelled

plt.figure(figsize=(5,4))
plt.title('Reservation Status Count')
plt.bar(['not cancelled','cancelled'],df['is_canceled'].value_counts(), edgecolor='k', width=0.7,color='#df5b47')
plt.show()
```

```python
#now lets see which hotel has more cancellation rate
plt.figure(figsize=(8,5))
ax1=sns.countplot(x='hotel',hue='is_canceled',data=df,palette='blend:#ff9b47,#bf1b47')
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status in Different Hotels')
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')
plt.show()
```

```python
#bringing the above chart down to percentages
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#the cancellation numbers are surging. lets see does price has any affect on the cancellation rates 
#of both these hotels
#we will do this by seeing the mean of the Average Daily Rate on a particular day of reservation
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
print(resort_hotel)
print(city_hotel)
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel',color='#ff9b47')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel',color='#bf1b47')
plt.legend(fontsize=20)
plt.show()
```

```python
#prices of city hotel are less than resort for some days
#The spikes that we are seeing in the graph happen during the weekend.

#lets see which month has more reservations and cancellations
#we are using a countplot again.
#we want to see a grouped bar chart on each month
df['month']=df['reservation_status_date'].dt.month #creating a new column called month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month',hue='is_canceled',data=df,palette='blend:#ff9b47,#bf1b47')
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status per month', fontsize=20)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.legend(['Not cancelled','Cancelled'])
plt.show()
```

```python
#in the above graph august has the most number of reservations and least cancellations
#september also has least cancellations
#january has the least reservations and more cancellations
#is it because the prices in august or september were low and in january or december the prices were high?

#plotting average daily rate for each month to see that
#we are only gonna use cancellation rates now to see the effect of price on it
plt.figure(figsize=(15,8))
plt.title('Average Daily Rate Per Mpnth', fontsize=30)
sns.barplot('month','adr', data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index(), palette='blend:#ff9b47,#bf1b47')
plt.show()
```

```python
#from the above graph we can see that august through december has some of the least adr
#and the highest prices are in january
#it proves our hypothesis that when price is higher the cancellation rates will be higher

#lets now see the cancellation rate based on top 10 countries

cancelled_data=df[df['is_canceled']==1]
countries=cancelled_data['country'].value_counts()[:10] #first 10 countries will be stored
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with reservation cancelled')
plt.pie(countries,autopct='%.2f',labels=countries.index,colors=["#ff9b47","#ef7b47","#df5b47",
                                                                "#cf3b47","#bf1b47","#9a0b56",
                                                                "#54045e","#03827e","#006b76",
                                                                "#0c4d5b"]) 
#autopct is here is used to show values till two decimal points
plt.show()
```

```python
#portugal has the most number of cancellations
#the hotels should either lessen their price or offer more services at the current price

#lets see from where the clients are coming
#are they booking online or coming from offline travel agents
#the hypothesis is that majority of clients are coming from offline travel agents

df['market_segment'].value_counts()
```

```python
df['market_segment'].value_counts(normalize=True)*100
```

```python
#the hypothesis is wrong. majority of clients are coming through online travel agents
#lets see from where the most cancellations are coming from
cancelled_data['market_segment'].value_counts(normalize=True)*100
```

```python
#why is this online cancellation rate more than others? 
#maybe the hotels are not putting up their true photos and are exaggerating the services they provide
#to minimize these cancellation rates the hotels must stay true to their image

#now lets see if adr of the cancelled reservations is high or non-cancelled reservation is high

cancelled_df_adr=cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date',inplace=True)


ncancelled_data=df[df['is_canceled']==0]
ncancelled_df_adr=ncancelled_data.groupby('reservation_status_date')[['adr']].mean()
ncancelled_df_adr.reset_index(inplace=True)
ncancelled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate', fontsize=30)
plt.plot(ncancelled_df_adr['reservation_status_date'],
         ncancelled_df_adr['adr'],label='Not Cancelled',color='#ff9b47')
plt.plot(cancelled_df_adr['reservation_status_date'],
         cancelled_df_adr['adr'],label='Cancelled',color='#bf1b47')
plt.legend(fontsize=20)
plt.show()
```

```python
#this graph is messy. 
#because we dont have consistent data before 2015 and after 2017.
#to make it visually better we'll plot the graph from 2016-2017

cancelled_df_adr=cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]
ncancelled_df_adr=ncancelled_df_adr[(ncancelled_df_adr['reservation_status_date']>'2016') & (ncancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate', fontsize=30)
plt.plot(ncancelled_df_adr['reservation_status_date'],
         ncancelled_df_adr['adr'],label='Not Cancelled',color='#ff9b47')
plt.plot(cancelled_df_adr['reservation_status_date'],
         cancelled_df_adr['adr'],label='Cancelled',color='#bf1b47')
plt.legend(fontsize=20)
plt.show()
```
